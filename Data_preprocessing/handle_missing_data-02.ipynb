<h1 id="identify_handle_missing_values">Identify and handle missing values</h1>

<h2 id="identify_missing_values">Identify missing values</h2>
<h3>Convert "?" to NaN</h3>
In the car dataset, missing data comes with the question mark "?".
We replace "?" with NaN (Not a Number), which is Python's default missing value marker, for reasons of computational speed and convenience. Here we use the function: 
 <pre>.replace(A, B, inplace = True) </pre>
to replace A by B


```python
df.replace('?', np.nan, inplace = True)
df.head(15)
```

```python
df.info()
```

Identify_missing_values

<h2>Evaluating for Missing Data</h2>

The missing values are converted to default. We use the following functions to identify these missing values. There are two methods to detect missing data:

<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.


```python
df.isnull().head(15)
```

"True" stands for missing value, while "False" stands for not missing value.

<h2>Count missing values in each column</h2>
<p>
Using a for loop in Python, we can quickly figure out the number of missing values in each column. As mentioned above, "True" represents a missing value, "False"  means the value is present in the dataset.  In the body of the for loop the method  ".value_counts()"  counts the number of "True" values. 
</p>


```python
df.isnull().sum()
```

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
missing_data = df.isnull()
for column in headers:
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

<h2 id="deal_missing_values">Deal with missing data</h2>
<br>
<h3>How to deal with missing data?</h3>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data the missing values<br>
        a. replace it by an average (Mean and Median)<br>
        b. replace it by frequency (Mode)<br>
    </li>
    <li> Leave it as missing data </li>
</ol>

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four". 
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
        </ul>
    </li>
</ul>


<h3>Calculate the average of the normalized-losses column </h3>


```python
df.dtypes
```

```python
avg_nor_loss = df["normalized-losses"].astype('float').mean(axis=0)
avg_nor_loss
```

<h3>Replace "NaN" by mean value in "normalized-losses" column</h3>


```python
df['normalized-losses'].replace(np.nan, avg_nor_loss, inplace = True)
df.head(15)
```

<h3>Calculate the mean value for 'bore' column</h3>

```python
bore_avg = df['bore'].astype('float').mean(axis = 0)
bore_avg
```

```python
df['bore'].replace(np.nan, bore_avg, inplace = True)
df.head(15)
```

<h3>Calculate the mean value for 'stroke' column</h3>
and
<h3>Replace NaN by mean value for stroke column</h3>


```python
stroke_avg = df['stroke'].astype('float').mean(axis=0)
df['stroke'].replace(np.nan, stroke_avg, inplace = True)
```

<h5>Calculate the mean value for 'peak-rpm' column:</h5>
and 
<h5>Replace NaN by mean value:</h5>


```python
peak_rpm_avg = df['peak-rpm'].astype(float).mean(axis=0)
df['peak-rpm'].replace(np.nan, peak_rpm_avg, inplace=True)
```

<h4>Calculate the mean value for the  'horsepower' column:</h4>
and
<h4>Replace "NaN" by mean value:</h4>

```python
horsepower_avg = df['horsepower'].astype(float).mean(axis=0)
df['horsepower'].replace(np.nan, horsepower_avg, inplace=True)
```

## Finally, let's drop all rows that do not have price data: dropna()

### inplace = true means save updated data frame and 
### axis = 0 means full row delete 

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=['price'], inplace=True, axis=0)

# reset index, because we dropped two rows
df.reset_index(drop=True, inplace=True)
df.head(20)
```
