A **Linear Regression problem** is one of the simplest and most fundamental problems in machine learning and statistics.

### 🔹 **Definition**

Linear regression is a **supervised learning algorithm** used to model the relationship between a **dependent variable (target)** and one or more **independent variables (features)** by fitting a straight line (in higher dimensions, a hyperplane) to the data.

### 🔹 **Goal**

The goal is to **predict the target variable** ( y ) based on the values of the input features ( x ).
We try to find a linear equation of the form:

[
y = m x + c
]
(for one variable)

or more generally,
[
y = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
]
(for multiple variables)

where:

* ( y ) = predicted output (target)
* ( x_i ) = input features
* ( w_i ) = model coefficients (weights)
* ( b ) = intercept (bias term)

### 🔹 **Objective**

The algorithm learns the values of ( w_i ) and ( b ) that **minimize the prediction error** — typically measured by the **Mean Squared Error (MSE)**:

[
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
]

where ( y_i ) is the actual value and ( \hat{y}_i ) is the predicted value.

### 🔹 **Example Problem**

Suppose you want to **predict house prices** based on size (square feet).

* Input (feature): size of the house
* Output (target): price of the house
  Linear regression learns the best-fitting line through the data that predicts price from size.

### 🔹 **Applications**

* Predicting sales, demand, or prices
* Forecasting trends
* Estimating relationships between variables
